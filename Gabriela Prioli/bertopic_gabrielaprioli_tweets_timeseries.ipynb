```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/silvinhad/Desktop/bertopic Python Analysis/gabrielaprioli_tweets.csv').astype("string")
data.head(3)
```

```python
data["created_at"] = pd.to_datetime(data["created_at"])
```

```python
ts_1h = (
    data
    .set_index("created_at")
    .resample("h")
    .size()
    .reset_index(name="count")
)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(14, 5))
plt.plot(ts_1h["created_at"], ts_1h["count"])
plt.xlabel("Date")
plt.ylabel("Number of tweets")
plt.title("Tweet Volume Per Hour Mentioning Gabriela Prioli")
plt.show()
```
